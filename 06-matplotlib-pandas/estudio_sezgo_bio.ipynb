# Estudio del Sezgo Educativo De Licencenciaturas en Biología por Género
> Nota: El ejercicio ha sido extraído del curso _Storytelling Through Data Visualization_ de [Dataquest](https://www.dataquest.io) 

El Departamento de Estadísticas de Educación de EEUU publica anualmente un conjunto de datos que contiene el porcentaje de títulos de licenciatura otorgados a mujeres de 1970 a 2012. El conjunto de datos se divide en 17 categorías de títulos, con cada columna como una categoría separada. Randal Olson, científico de datos de la Universidad de Pensilvania, limpió el conjunto de datos y lo puso a disposición en su sitio web personal. Puede descargar el conjunto de datos Randal compilado [aquí](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). A continuación se muestran las primeras 5 filas del conjunto de datos:

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
women_degrees = pd.read_csv('./examples/percent-bachelors-degrees-women-usa.csv')
women_degrees.head()
women_degrees.describe()
```

## Licenciaturas otorgadas a mujeres en biología
Vamos a ver .....

```python
fig, ax = plt.subplots()
ax.plot(women_degrees["Year"], women_degrees["Biology"], label="Mujeres")
ax.plot(women_degrees["Year"], 100-women_degrees["Biology"], label="Hombres")
ax.set_title("Porcentaje de Licenciaturas por Género", size=14)
#ax.tick_params(bottom="off", top="off", left="off", right="off")
for key,_ in ax.spines.items():
    ax.spines[key].set_visible(False)
ax.legend(loc="lower right")

plt.show()
```

Podemos notar que la tendencia cambió.

```python
majors = ['Agriculture', 'Architecture', 'Foreign Languages', 'Business']
fig = plt.figure(figsize=(12,12))
fig.suptitle("Sezgo de Licenciaturas Otorgadas por Género", size=20)
fig.tight_layout()
for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.scatter(women_degrees["Year"], women_degrees[majors[sp]], label="Mujeres")
    ax.scatter(women_degrees["Year"], 100-women_degrees[majors[sp]], label="Hombres")
    for key,_ in ax.spines.items():
        ax.spines[key].set_visible(False)
    ax.set_title(majors[sp])
    ax.set_ylim(0,100)

ax.legend();
```
