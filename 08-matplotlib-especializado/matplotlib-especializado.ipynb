## 1. Escogiendo la Visualización Apropiada
Tenemos una gran cantidad de visualizaciones y opciones a nuestra dispocisión, por lo que es importante escoger la que comunique de manera más clara e intuitiva la información presentada. A continuación, veremos unas clasificaciones generales que nos ayudarán a seleccionar la más adecuada según lo que deseemos mostrar.  
El material está inspirado en contenido realizado por [_edraw_](https://www.edrawsoft.com/chart/choose-right-chart.php).  
Otra fuente muy útil para escoger gráficas es el sitio [_The Python Graph Gallery_](https://python-graph-gallery.com/)

### 1.A. Presentar Distribución
Las gráficas de distribución muestran cómo los objetos de estudio se _distribuyen_ en diferentes partes. Las gráficas de esta categoría son gráficas de líneas, histogramas, gráficas de dispersión, entre otras.

### 1.B. Visualización de Composición
Este tipo de gráficas, presentan la información como parte de un entero. Las gráficas más apropiadas para esta categoría son las gráficas de pastel, gráficas de área y gráficas de barra apiladas con diferentes colores.

### 1.C. Mostrar Relación
En situaciones cuando se requiere mostrar la relación entre una variable y todo el grupo de las variables estudiadas, utilizamos gráficas radiales o de burbujas.

### 1.D. Indicar Tendencia
Cuando requerimos indicar la tendencia de una serie de datos en un periodo fijo, las dos gráficas más apropiadas son gráficas de líneas y gráficas de columnas. Ambas gráficas muestran los cambios de las variables en una forma intuitiva y fácilmente comparable. También se pueden combinar en una gráfica de _Paretto_.

## Graficando Diferentes Visualizaciones en `matplotlib`
A continuación, estudiaremos los métodos utilizados para realizar diferentes tipos de gráficas con Python y `matploltib`:

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
women = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women.head()
```

### 2.A. Gráficas de Columnas
Para realizar una gráfica de columnas utilizamos el método [`plt.bar()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html) de la siguiente forma:

```python
plt.style.use('seaborn-poster')
plt.style.use('bmh')
fig, ax1 = plt.subplots()
ax1.bar(women['Year'], women['Architecture'])
ax1.set_ylim(0,100)
plt.show()
```

### 2.B. Gráficas de Barras Horizontales
Para este tipo de gráficas utilizamos el método [`plt.barh()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.barh.html) de la siguiente forma:

```python
np.random.seed(19680801)

people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

fig, ax = plt.subplots()
ax.barh(people, performance, xerr=error)
ax.invert_yaxis()
ax.set_xlabel("Performance")
plt.show()
```

### 3.C. Histogramas
**Nota:** Un histograma y una diagrama de barras/columnas **NO** son lo mismo. Para mayor explicación consultar [este link.](https://anestesiar.org/2014/como-un-huevo-una-castana-diferencias-entre-un-diagrama-de-barras-y-un-histograma/)  
Para graficar histogramas utilizamos el método [`plt.hist()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html) de la siguiente forma:

```python
# datos de ejemplo
mu = 100  # media de la distribución
sigma = 15  # desviación estándar de la distribución
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots()

# histograma de los datos
n, bins, patches = ax.hist(x, num_bins, density=1)

ax.set_xlabel('IQ')
ax.set_ylabel('Densidad de la Probabilidad')
ax.set_title(r'Histograma de IQ: $\mu=100$, $\sigma=15$')

fig.tight_layout()
plt.show()
```

### 3.D. Gráficas de Dispersión
Para realizar gráficas de dispersión utilizamos el método [`plt.scatter()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html) de la siguiente forma:

```python
fig, ax = plt.subplots()

# datos de ejemplo
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
color = np.random.rand(N)

# gráfica de dispersión
ax.scatter(x, y, c=color)
plt.show()
```

### 3.E. Gráficas de Pastel
Para realizar una gráfica de pastel utilizamos el método [`plt.pie()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.pie.html) de la siguiente forma:

```python
# diagrama de paste, las 'rebanadas' serán ordenadas y graficadas en sentido contrario a las manecillas del reloj:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # sólo separar la 2a rebanada

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Radio de aspecto 'equal' garantiza que la gráfica se dibuje como círculo

plt.show()
```

### 3.F. Gráficas de Área
Para realizar gráficas de área utilizamos el método [`plt.fill_between()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.fill_between.html) de la siguiente forma.

```python
fig, (ax1, ax2) = plt.subplots(ncols=1, nrows=2)
# datos de ejemplo
x= np.arange(1,15)
y= np.random.rand(len(x))
 
# Gráfica de Área básica
ax1.fill_between( x, y)
 
# Same, but add a stronger line on top (edge)
ax2.fill_between( x, y, color="skyblue", alpha=0.2)
ax2.plot(x, y, color="Slateblue", alpha=0.6)

plt.show()
```

### 3.G. Diagramas de Caja
Para realizar gráficas de caja utilizamos el método [`pd.DataFrame.boxplot()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html) de la siguiente forma

```python
lombrices = pd.read_excel('lombrices.xlsx')
fig, ax = plt.subplots()
ax = lombrices.boxplot()
plt.show()
```
